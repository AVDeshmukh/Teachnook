```python
#Slicing - Printing Sequence of data

a="randomness"
print(a[0:6])
print(a[0])
print(a[-2])
print(a[3:])
print(a[:6])
print(a[-8:-1])
print(a[-10:])
#step index
print(a[0:5:2]) #start from 0 to 5 printing by distance of 2
print(a[0:6:1]) #default it is 1
print(a[::-2])
```

```python

#keyword arguments
def key(name,subject,marks):
  print("my name",name, "my subject",subject,"marks",marks)
key(name="ad",subject="Math",marks=95)


#positional argument
def key(name,subject,marks):
  print("my name",name, "my subject",subject,"marks",marks)
key("ad","Math",95)

#default argument
def awesome(name,city="AMT",describe="beautiful"):
  print("hey i am",name,"from",city,"city is",describe)
awesome("tom","nagpur","Orange")
```

```python
def func(teacher,*a,**kwargs):
  print("teacher's name is",teacher)
  for roll in a:
    print("The students are",roll)
  for subject in kwargs:
    print(subject,kwargs[subject])
func("Tom","ad",1,2,3,science=10,eng=4)
```
